```python
#Trignometric functions
#Sin(x)
#Cos(x)
#Tan(x)-Sin(x)/Cos(x)
#Cosec-1/Sin(x)
#Sec-1/Cos(x)
#cot-1/Tan(x)
#
angles=np.array([0,15,31,45,60,75,90])
#Sin(array)
result=np.sin(np.radians(angles))
result1=np.cos(np.radians(angles))
result2=np.tan(np.radians(angles))
```

```python
print(result)
print(result1)
print(result2)
```

```python
arr=np.array([1,2,3,4,5,6])
#Reshape the array
reshapedarr=arr.reshape(3,2)
print(reshapedarr)
#cover/Faltten any n dim array
flattendarr=arr.flatten()
print(flattendarr)
```

```python
#stacking the data

arr1=np.array([1,2,3,4])
arr2=np.array([10,20,30,40])

#vertical stacking
arr3=np.vstack((arr1,arr2))
print(f"The shape of arr3 is{arr3.shape}")


#Horizontalstacking
arr4=np.vstack((arr1,arr2))
print(f"The shape of arr4 is{arr3.shape}")
```

```python
#Filtering
arr=np.array([10,20,30,40,50,60,70,80,90])
farr=arr[arr>40]
print(farr)
```
